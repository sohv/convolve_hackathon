```python
cols_to_transform = data.loc[:, ~data.columns.isin(['account_number', 'bad_flag'])].columns
```

```python
data[cols_to_transform] = np.log1p(data[cols_to_transform]+1)
```

```python
print(data.skew())
```

```python
print('Pearson Correlations between each column and the bad_flag')
zero_corr_cols=[]
nan_count=0
for i in data.columns:
  x=data["bad_flag"].corr(data[i])
  print(f'{i}: {x}')
  if pd.isna(x):
    nan_count+=1
    zero_corr_cols.append(i)
```

```python
print(f'Total nan correlations: {nan_count}')
```

```python
zero_corr_cols
```

```python
pearson_data = data.copy()
pearson_data.shape
```

```python
columns_to_drop = data.columns[data.columns.isin(zero_corr_cols)]
pearson_data = data.drop(columns=columns_to_drop)
```

```python
from scipy.stats import kendalltau
result=[]
for i in data.columns:
    if i!='account_number' or i!='bad_flag':
      tau,p_value=kendalltau(data["bad_flag"],data[i])
      result.append({"Column": i, "tau": tau, "p-value": p_value})
```

```python
kt=pd.DataFrame(result)
```

```python
kt.isnull().sum()
```

```python
kt.describe()
```

```python
# bad features acc to the kendall-tau correlation
kendalltau_insignificant_features=[]
for col in data.columns:
  tau,p_value=kendalltau(data["bad_flag"],data[col])
  if p_value>0.05 and -0.1<tau<0.1 :
    kendalltau_insignificant_features.append(col)
```

```python
kendalltau_insignificant_features.remove('account_number')
```
